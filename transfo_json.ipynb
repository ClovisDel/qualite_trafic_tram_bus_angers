```python
from google.colab import drive
from google.colab import files
drive.mount('/content/gdrive')

import pandas as pd
import numpy as np 
import json
from pandas import json_normalize

from tqdm import tqdm 

import matplotlib.pyplot as plt

import seaborn as sns

import time

import os 
```

```python
directory_path = '/content/gdrive/MyDrive/Colab Notebooks/Qualite_graphique/angers-bus-trams'
directory_files = os.listdir(directory_path)
```

```python
df = pd.DataFrame()
for i in tqdm(range(len(directory_files))):
  try :
    file_path = directory_path + "/" + directory_files[i]
    with open(file_path) as data_file:
      data = json.load(data_file)
      df_data = json_normalize(data["records"])
  except:
    pass

  df = pd.concat([df, df_data], axis=0, ignore_index=True)
```
