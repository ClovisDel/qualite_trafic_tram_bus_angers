```python
# plotly boxplot of ecart_horaire_en_secondes by OPINION with beeswarm

fig = px.box(df, x="OPINION", y="ecart_horaire_en_secondes", color="OPINION", points="all", hover_data=df.columns)
fig.update_layout(title_text='Boxplot of ecart_horaire_en_secondes by OPINION with beeswarm')
fig.show()
```
